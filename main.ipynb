```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
merge = pd.read_csv('./results/merge_0.5.csv', index_col=False)
heap = pd.read_csv('./results/heap_0.5.csv', index_col=False)
quick = pd.read_csv('./results/quick_0.5.csv', index_col=False)
quick_med = pd.read_csv('./results/quick med_0.5.csv', index_col=False)
quick_rand = pd.read_csv('./results/quick random_0.5.csv', index_col=False)
merge
```

```python
def plots(df, name):
   
   coefficients = np.polyfit(np.log(df['Size']),df['Time'],1) # Use log(x) as the input to polyfit.
   fit = np.poly1d(coefficients) 
   display(fit(np.log(df['Size'])))
   
   plt.plot(df['Size'], df['Time'], 'o', label='data')
   plt.plot(df['Size'],fit(np.log(df['Size'])),"--", label="fit")
   plt.title(name)
   plt.xlabel("Size")
   plt.ylabel("Time")  
   plt.legend()
```

```python
plots(merge, "Merge")
```

```python
plots(heap, "Heap")
```

```python
plots(quick, "Quick")
```

```python
plots(quick_rand, "Quick Rand")
```

```python
plots(quick_med, "Quick Med")
```
