```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
merge_100 = pd.read_csv('./results/merge_100.csv', index_col=False)
merge_1000 = pd.read_csv('./results/merge_1000.csv', index_col=False)
merge_5000 = pd.read_csv('./results/merge_5000.csv', index_col=False)
merge_10000 = pd.read_csv('./results/merge_10000.csv', index_col=False)
merge_50000 = pd.read_csv('./results/merge_50000.csv', index_col=False)
merge_100000 = pd.read_csv('./results/merge_100000.csv', index_col=False)

merge = [merge_100, merge_1000, merge_5000, merge_10000, merge_50000, merge_100000]
```

```python
heap_100 = pd.read_csv('./results/heap_100.csv', index_col=False)
heap_1000 = pd.read_csv('./results/heap_1000.csv', index_col=False)
heap_5000 = pd.read_csv('./results/heap_5000.csv', index_col=False)
heap_10000 = pd.read_csv('./results/heap_10000.csv', index_col=False)
heap_50000 = pd.read_csv('./results/heap_50000.csv', index_col=False)
heap_100000 = pd.read_csv('./results/heap_100000.csv', index_col=False)

heap = [heap_100, heap_1000, heap_5000, heap_10000, heap_50000, heap_100000]
```

```python
result_merge = []
for element in merge:
   result_merge.append([element['Size'].mean(), element['Time'].mean()])
   
result_merge_df = pd.DataFrame(result_merge, columns=['Size', 'Time'])
```

```python
result_heap = []
for element in heap:
   result_heap.append([element['Size'].mean(), element['Time'].mean()])
   
result_heap_df = pd.DataFrame(result_heap, columns=['Size', 'Time'])
```

```python
n = result_merge_df['Size']
x = result_merge_df['Time']
plt.plot(result_merge_df['Time'], n, 'o')
plt.title("Merge Sort")
plt.xlabel("Mean Time")
plt.ylabel("Sizes")
```

```python
n = result_heap_df['Size']
size = n*np.log2(n)
plt.plot(result_heap_df['Time'], n)
plt.title("Heap Sort")
plt.xlabel("Mean Time")
plt.ylabel("Sizes")
```
